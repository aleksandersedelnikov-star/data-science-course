##  Анализатор успеваемости
### Задание:
#### 1. Запросите у пользователя 5 оценок за экзамены (от 2 до 5)
#### 2. Рассчитайте:
##### - Средний балл
##### - Максимальную оценку
##### - Минимальную оценку
##### - Количество пятерок, четверок, троек и двоек
#### 3. Определите статус студента:
##### - Отличник (все оценки 5)
##### - Хорошист (есть 4 и 5, нет 2 и 3)
##### - Удовлетворительно (есть 3)
##### - Неудовлетворительно (есть 2)
#### 4. Выведите развернутый отчет

```python
count_five = 0
count_four= 0
count_three = 0
count_two = 0
sum_value = 0
min_value = 0
max_value = 0
i = 0
try:
    # Запрашиваем ввод оценок и заполняем переменные/попутно проверяем минимальное и максимальное значение
    while (i < 5):
        buff_value = int(input(f"Введите {i+1} оценку за экзамен (диапазон значений [2,5]): "))
        if (i == 0):
            min_value, max_value = buff_value, buff_value
        else:
            if (buff_value < min_value):
                min_value = buff_value
            elif (buff_value > max_value):
                max_value = buff_value
        if (buff_value == 2):
            count_two += 1
        elif (buff_value == 3):
            count_three += 1
        elif (buff_value == 4):
            count_four += 1
        elif (buff_value == 5):
            count_five += 1
        else:
            print("Введено некорректное значение (не в диапазоне от 2 до 5).")
            continue
        sum_value += buff_value
        i += 1
            
    # Рассчитываем средний балл
    average = sum_value / 5

    # Определяем тип студента
    if (count_two == 0 and count_three == 0 and count_four == 0):
        type_student = 'Отличник'
    elif (count_two == 0 and count_three == 0 and count_four != 0):
        type_student = 'Хорошист'
    elif (count_two == 0 and count_three != 0):
        type_student = 'Удовлетворительно'
    elif (count_two != 0):
        type_student = 'Неудовлетворительно'
    else:
        type_student = 'Неопределено'
        
    # Выводим отчет
    print("\n\n= = = Отчет по введенным оценкам: = = =")
    print(f"Средний балл студента: {average}")
    print(f"Максимальная оценка: {max_value}")
    print(f"Минимальная оценка: {min_value}")
    print(f"Количество 5-ок: {count_five} шт., Количество 4-ок: {count_four} шт., Количество 3-ек: {count_three} шт., Количество 2-ек: {count_two} шт.")
    print(f"Тип студента: {type_student}")
    print("= = = = = = = = = = = = = = = = = = = =")
except Exception as e:
    print(f"Произошла ошибка: {e}")
```
